```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Drowsy_mear.csv')
#df1.columns = ["ear", "mar"]
#df1.head()
df1 = df1.T
df1.head()
```

```python
df1.reset_index(drop=True, inplace=True)
df1.head()
```

```python
df1['ear'] = df1[0]
df1['mar'] = df1[1]
df1.drop(0, axis=1, inplace=True)
df1.drop(1, axis=1, inplace=True)
df1.head()
```

```python
df1.tail(20)
df1.to_csv('drowsy.csv', index=False)
```

```python
df1.info()
```

```python
plt.figure(figsize=(25,25))
plt.scatter(df1['mar'], df1['ear'], color='red')
plt.show()
```

```python
df2 = pd.read_csv('NotDrowsy_mear.csv')
df2 = df2.T
df2.head()
```

```python
df2.reset_index(drop=True, inplace=True)
df2['ear'] = df2[0]
df2['mar'] = df2[1]
df2.drop(0, axis=1, inplace=True)
df2.drop(1, axis=1, inplace=True)
df2.head()
```

```python
df2.to_csv('Ndrowsy.csv', index=False)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df2['mar'], df2['ear'], color='green')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df1['mar'], df1['ear'], color='red')
plt.scatter(df2['mar'], df2['ear'], color='green')
plt.xlabel('MAR')
plt.ylabel('EAR')
plt.show()
```

```python
plt.figure(figsize=(25,25))
plt.scatter(df1['ear'], df1['mar'], color='red')
plt.scatter(df2['ear'], df2['mar'], color='green')
plt.show()
```
