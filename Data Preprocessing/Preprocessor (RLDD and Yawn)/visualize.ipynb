```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Person 3/drowsy_cleaned_P3_v2.csv')
df2 = pd.read_csv('Person 3/notdrowsy_cleaned_P3_v2.csv')
```

```python
df1.head(20)
```

```python
df2.head(20)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df1['mar'], df1['ear'], color='red',label='Drowsy')
plt.scatter(df2['mar'], df2['ear'], color='green', label='Not Drowsy')
plt.xlabel('MAR')
plt.ylabel('EAR')
plt.legend()
plt.savefig('Person 3/p3cv2.jpg',bbox_inches='tight')
plt.show()
```
