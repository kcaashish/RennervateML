```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Person 3/drowsy_P3.csv')
df.head()
```

```python
df.iloc[:,1:].values
```

```python
import math
ear=df.iloc[:,0:1].values
mar = df.iloc[:,1:].values
len(ear), len(mar)

```

```python
n = math.ceil(len(ear)/10)
n
```

```python
ear_a = np.array_split(ear, n)
mar_a = np.array_split(mar, n)
```

```python
ear_c = []
for i in ear_a:
    ear_i = np.amin(i)
    ear_c.append(ear_i)
    
print(ear_c)
len(ear_c)
```

```python
mar_c = []
for i in mar_a:
    mar_i = np.amax(i)
    mar_c.append(mar_i)
    
print(mar_c)
len(mar_c)
```

```python
type(mar_c)
```

```python
dc = pd.DataFrame(
    {'ear': ear_c,
     'mar': mar_c,
    })
```

```python
dc.head()
```

```python
dc.shape
```

```python
dc.to_csv('Person 3/drowsy_cleaned_P3.csv', index=False)
```
